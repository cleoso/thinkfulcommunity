# Data Visualization

```python
# To keep the page organized, do all imports here
from sqlalchemy import create_engine
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns 
sns.set()

# Suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")

project_df = pd.read_csv('DA_-_houseprices.csv')

# Add your code here
```

```python
# Determine general information of the data set
project_df.info()
```

```python
# Determine the head of the data table
project_df.head()
```

```python
# Determine the distribution of salesprice is skewed using a box plot
sns.boxplot(y="saleprice", data=project_df)
```

```python
# Determine lotarea and see there outliers above a lot area of 50,000. <=50K
sns.boxplot(y="lotarea", data=project_df)
```

```python
# Determine overall cond affects the sale price
sns.boxplot(y='overallcond', data=project_df)
```

```python
# Determine the sales price by overall condition and highlight the difference having central air conditioning or not
sns.boxplot(y="saleprice", hue='centralair', x='overallcond', data=project_df)
```

```python
# Determine for correlation between lotarea and saleprice
sns.scatterplot(y="lotarea", x='saleprice', data=project_df)
```
